```python
# ===============================================
# 3D Object Reconstruction from 2D Images
# ===============================================
# Works in Google Colab - no API keys needed
# Libraries: OpenCV, PyTorch, MiDaS, Open3D
# ===============================================

# Install dependencies
!pip install torch torchvision torchaudio --quiet
!pip install opencv-python open3d matplotlib timm --quiet

import cv2
import torch
import urllib
import matplotlib.pyplot as plt
import numpy as np
import open3d as o3d
from torchvision.transforms import Compose, Resize, ToTensor
from PIL import Image

# ------------------------------------------------
# 1️⃣ Load MiDaS depth estimation model
# ------------------------------------------------
midas = torch.hub.load("intel-isl/MiDaS", "DPT_Large")
midas_transforms = torch.hub.load("intel-isl/MiDaS", "transforms")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
midas.to(device)
midas.eval()

transform = midas_transforms.dpt_transform

# ------------------------------------------------
# 2️⃣ Load an input 2D image (you can replace the URL or upload one)
# ------------------------------------------------
# ------------------------------------------------
# 2️⃣ Load an input 2D image
# ------------------------------------------------

# Option A: Use a different public image URL
url = "https://raw.githubusercontent.com/opencv/opencv/master/samples/data/lena.jpg"
urllib.request.urlretrieve(url, "input.jpg")

# Option B: (Recommended) Upload your own image
# from google.colab import files
# uploaded = files.upload()
# img_path = list(uploaded.keys())[0]
# img = cv2.imread(img_path)[:, :, ::-1]

img = cv2.imread("input.jpg")[:, :, ::-1]
plt.imshow(img)
plt.title("Input 2D Image")
plt.axis('off')
plt.show()


# ------------------------------------------------
# 3️⃣ Estimate depth map
# ------------------------------------------------
input_batch = transform(img).to(device)
with torch.no_grad():
    prediction = midas(input_batch)
    prediction = torch.nn.functional.interpolate(
        prediction.unsqueeze(1),
        size=img.shape[:2],
        mode="bicubic",
        align_corners=False,
    ).squeeze()

depth_map = prediction.cpu().numpy()

plt.imshow(depth_map, cmap='plasma')
plt.title("Estimated Depth Map")
plt.axis('off')
plt.show()

# ------------------------------------------------
# 4️⃣ Convert depth map → 3D point cloud
# ------------------------------------------------
h, w = depth_map.shape
focal = w  # approximate focal length

Q = np.float32([[1, 0, 0, -w/2],
                [0, -1, 0, h/2],
                [0, 0, 0, -focal],
                [0, 0, 1, 0]])

points_3D = cv2.reprojectImageTo3D(depth_map, Q)
colors = img.reshape(-1, 3) / 255.0
mask = depth_map > depth_map.min()
output_points = points_3D[mask]
output_colors = colors[mask.flatten()]

# ------------------------------------------------
# 5️⃣ Visualize 3D point cloud
# ------------------------------------------------
pcd = o3d.geometry.PointCloud()
pcd.points = o3d.utility.Vector3dVector(output_points)
pcd.colors = o3d.utility.Vector3dVector(output_colors)

print("✅ 3D Reconstruction Successful! Displaying Point Cloud...")

# Save and show the point cloud
o3d.io.write_point_cloud("reconstructed_pointcloud.ply", pcd)
o3d.visualization.draw_geometries([pcd])

# ------------------------------------------------
# 6️⃣ (Optional) Save all results
# ------------------------------------------------
cv2.imwrite("depth_map.png", (depth_map / depth_map.max() * 255).astype(np.uint8))
print("Saved: input.jpg, depth_map.png, reconstructed_pointcloud.ply")
```

```python
!pip install gradio numpy opencv-python matplotlib scikit-image open3d

import gradio as gr
import numpy as np
import cv2
import open3d as o3d
from matplotlib import pyplot as plt
from skimage import filters
from PIL import Image

def reconstruct_3d(image):
    if isinstance(image, np.ndarray):
        image = Image.fromarray(image.astype('uint8')).convert("RGB")

    img = np.array(image)
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

    # Generate a fake depth map (for demo)
    depth = filters.sobel(gray)

    # Create simple 3D point cloud
    h, w = depth.shape
    xx, yy = np.meshgrid(np.arange(w), np.arange(h))
    zz = depth * 255

    points = np.stack((xx, yy, zz), axis=-1).reshape(-1, 3)
    colors = img.reshape(-1, 3) / 255.0

    pc = o3d.geometry.PointCloud()
    pc.points = o3d.utility.Vector3dVector(points)
    pc.colors = o3d.utility.Vector3dVector(colors)

    o3d.io.write_point_cloud("reconstruction.ply", pc)

    # Display depth map
    plt.imshow(depth, cmap='plasma')
    plt.axis('off')
    plt.savefig("depth_map.png", bbox_inches='tight', pad_inches=0)
    plt.close()

    return "depth_map.png", "reconstruction.ply"

iface = gr.Interface(
    fn=reconstruct_3d,
    inputs=gr.Image(type="numpy", label="Upload 2D Image"),
    outputs=[
        gr.Image(label="Depth Map"),
        gr.Model3D(label="3D Reconstruction")
    ],
    title="3D Object Reconstruction from 2D Images",
    description="Upload a single 2D image to reconstruct its approximate 3D shape.",
    allow_flagging="never"
)

iface.launch(share=True)
```
